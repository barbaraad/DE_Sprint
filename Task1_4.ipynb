```python
import numpy as np
import matplotlib.pyplot as plt
```

#Датасет

```python
age=[]
sex=[]
imt=[]
chil=[]
smoker=[]
region=[]
insure=[]
```

```python
with open("/content/dataset_home.txt", mode="r") as file:
  for human in file.readlines():
    age.append(int(human.split(",")[0]))
    if human.split(",")[1]=="male":
      sex.append(1)
    else:
      sex.append(0)
    imt.append(float(human.split(",")[2]))
    chil.append(int(human.split(",")[3]))
    if human.split(",")[4]=="yes":
      sex.append(1)
    else:
      sex.append(0)
    if human.split(",")[5]=="northwest":
      region.append(0)
    elif human.split(",")[5]=="northeast":
      region.append(1)
    elif human.split(",")[5]=="southwest":
      region.append(2)
    elif human.split(",")[5]=="southeast":
      region.append(3)
    else: region.append(human.split(",")[5])
    insure.append(float(human.split(",")[6]))
```

```python
age=np.array(age)
sex=np.array(sex)
imt=np.array(imt)
chil=np.array(chil)
smoker=np.array(smoker)
region=np.array(region)
insure=np.array(insure)
```

Распределние Рэлэя

```python
import math
from scipy.stats import rayleigh
def ray(x:float)->float:
  return  (x/p)*np.exp(-(x**2)/2*p)
rv=rayleigh.rvs(size=100)
p=rv.var()
plt.plot(rv,[ray(x) for x in rv])
plt.hist(rv, color='lightblue')
print(p)
print(rv)
```

```python
def grafik1(x:float)->float:
  return  0.5*(((x**2)**(1/3))+((x**4)*(1/3)-4*(x**2-1))**(0.5)) 
def grafik2(x:float)->float:
  return  0.5*(((x**2)**(1/3))-((x**4)*(1/3)-4*(x**2-1))**(0.5))
xg=np.arange(-1.0492,1.0492,0.00001)
heart1=plt.plot(xg,[grafik1(x) for x in xg],'r')
heart2=plt.plot(xg,[grafik2(x) for x in xg],'b')
plt.xlim([-2,2])
plt.ylim([-2,2])


```
