```python
import numpy as np
import matplotlib.pyplot as plt
```

#Датасет

```python
age=[]
sex=[]
imt=[]
chil=[]
smoker=[]
region=[]
insure=[]
```

```python
with open("/content/dataset_home.txt", mode="r") as file:
  for human in file.readlines():
    age.append(int(human.split(",")[0]))
    if human.split(",")[1]=="male":
      sex.append(1)
    else:
      sex.append(0)
    imt.append(float(human.split(",")[2]))
    chil.append(int(human.split(",")[3]))
    if human.split(",")[4]=="yes":
      smoker.append(1)
    else:
      smoker.append(0)
    if human.split(",")[5]=="northwest":
      region.append(0)
    elif human.split(",")[5]=="northeast":
      region.append(1)
    elif human.split(",")[5]=="southwest":
      region.append(2)
    elif human.split(",")[5]=="southeast":
      region.append(3)
    else: region.append(human.split(",")[5])
    insure.append(float(human.split(",")[6]))
```

```python
age=np.array(age)
sex=np.array(sex)
imt=np.array(imt)
chil=np.array(chil)
smoker=np.array(smoker)
region=np.array(region)
insure=np.array(insure)
```

```python
plt.scatter(age[sex==0],imt[sex==0])
plt.scatter(age[sex==1],imt[sex==1])
```

Индекс массы тела и возраст независимые величины, причем как для мужчин, так и для женщин. От пола распределение данных величин не зависит.

```python
plt.scatter(imt[smoker==0],insure[smoker==0])
plt.scatter(imt[smoker==1],insure[smoker==1])
```

Курящие тратят на страховку больше средств, чем не курящие. Кроме того, если среди некурящих в целом наблюдается равномерное распределение зависимости затрат на страховку от ИМТ, то среди курящих объем затрат на страховку прямо пропорционален ИМТ.

```python
plt.scatter(age[smoker==1],insure[smoker==1])
plt.scatter(age[smoker==0],insure[smoker==0])
```

В целом, сумма затраченная на страховку пропорциональна возрасту, хотя имеется две грyппы людей, которые тратят на страховку в 2 и в 4 раза больше среднестатистических значений, скорее всего, это люди с более высоким уровнем жизни. также отчетливо прослеживается тенденция, чтот курящие тратят на страховку гораздо больше не курящих.
